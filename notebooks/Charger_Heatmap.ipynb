```python
# New dataset test
df = pd.read_csv('electric_stations.csv')
```

```python
NE_MIN_LAT, NE_MAX_LAT = 38.5, 47.5
NE_MIN_LON, NE_MAX_LON = -80.5, -66.8
```

```python
df = df[
    (df['Latitude'] >= NE_MIN_LAT) &
    (df['Latitude'] <= NE_MAX_LAT) &
    (df['Longitude'] >= NE_MIN_LON) &
    (df['Longitude'] <= NE_MAX_LON)
]
```

```python
df
```

```python
from folium.plugins import HeatMap

# Extract list of [lat, lon] pairs
heat_data = df[['Latitude', 'Longitude']].dropna().values.tolist()

# Create map centered on the mean location
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=6)

# Add heatmap layer
HeatMap(heat_data).add_to(m)

# Display
m
```
