```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('S1901.csv')
```

```python
df
```

```python
# Store the first row (descriptive headers)
header_row = df.iloc[0]

# Generate new column names by combining original column names with the descriptive headers
new_column_names = []
columns_to_drop_at_start = [] # For columns like 'Unnamed: 130'
for col in df.columns:
    desc_value = header_row[col]
    if pd.isna(desc_value) or not isinstance(desc_value, str):
        # If descriptive value is NaN or not a string, just use the original column name
        columns_to_drop_at_start.append(col) # Mark for dropping the original column name
        new_column_names.append(col) # Keep original name for now for df.columns assignment
    else:
        # Combine original column name with the descriptive value
        new_column_names.append(f"{desc_value.replace('!',' ')}")

# Assign the new column names to the DataFrame
df.columns = new_column_names

# Identify columns to drop based on new column names
moe_cols = [col for col in df.columns if 'Margin of Error' in col]
pa_cols = [col for col in df.columns if 'PERCENT ALLOCATED' in col]
dol_cols = [col for col in df.columns if '$' in col]

# Combine all columns to drop into a single list, ensuring no duplicates
# This correctly handles columns that are both 'Margin of Error' and 'PERCENT ALLOCATED'
all_cols_to_drop = list(set(moe_cols + pa_cols + dol_cols + columns_to_drop_at_start))

# Drop all identified columns in a single operation
df = df.drop(columns=all_cols_to_drop)

# Drop the first row (which was the descriptive header row)
df = df.iloc[1:].copy()

display(df.head())
```

```python
df['Zip_Code'] = df['Geographic Area Name'].str.extract(r'ZCTA5 (\d{5})')

display(df.head())
```

```python
df = df.drop(columns=['Geography', 'Geographic Area Name'])

# Get all columns except 'Zip_Code'
other_cols = [col for col in df.columns if col != 'Zip_Code']

# Create the new column order with 'Zip_Code' first
new_column_order = ['Zip_Code'] + other_cols

# Reindex the DataFrame with the new column order
df = df[new_column_order]

display(df.head())
```
