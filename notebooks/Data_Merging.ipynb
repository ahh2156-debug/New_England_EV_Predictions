```python
age_df = age.rename(columns={'zip': 'zip_code'})
if 'Unnamed: 0' in age_df.columns:
    age_df = age_df.drop(columns=['Unnamed: 0'])

employment_df = employment.rename(columns={'Zip_code': 'zip_code'})
if 'Unnamed: 0' in employment_df.columns:
    employment_df = employment_df.drop(columns=['Unnamed: 0'])

income_df = income.rename(columns={'Zip_Code': 'zip_code'})
if 'Unnamed: 0' in income_df.columns:
    income_df = income_df.drop(columns=['Unnamed: 0'])

housing_df = housing.rename(columns={'Zip_code': 'zip_code'})
if 'Unnamed: 0' in housing_df.columns:
    housing_df = housing_df.drop(columns=['Unnamed: 0'])

# housing_grouped already has zip_code as index, convert to column
#housing_grouped_df = housing_grouped.reset_index()

# Merge dataframes
df = pd.merge(age_df, employment_df, on='zip_code', how='outer')
df = pd.merge(df, housing_df, on='zip_code', how='inner')
df = pd.merge(df, income_df, on='zip_code', how='outer')

# Clean column names: replace '!!' with space and '!' with empty string
df.columns = df.columns.str.replace('!!', ' ', regex=False)
df.columns = df.columns.str.replace('!', '', regex=False)

print("Shape of the final merged dataframe:", df.shape)
display(df.head())

import numpy as np

df = df.replace("-", np.nan)
df = df.replace("N", np.nan)
df = df.replace("250,000+", 250000)
df = df.replace("250,000-", 250000)
```

```python
# Clean Dataframe file
nan_counts = df.isna().sum(axis=1)
df = df[nan_counts < 50]
```

```python
df = df.drop(columns=['Estimate  Married-couple families  Mean income (dollars)'])
```

```python
nan_counts = df.isna().sum(axis=1)
nan_counts.value_counts()
```

```python
for col in df.columns:
  if col == "zip_code":
    continue
  else:
    col_mean = df[col].astype(float).mean()
    df[col].fillna(col_mean, inplace=True)
```

```python
df = df.replace("2,500-", 2500)
```

```python
# Get NY Zipcodes
ny_zips = pd.read_csv('/content/drive/MyDrive/IEOR 4212 Project/Datasets/uszips.csv')
ny_zips = ny_zips[ny_zips['state_id'] == 'NY']
ny_zips
```

```python
df['zip_code'] = df['zip_code'].astype(str).str.strip().str.zfill(5)
ny_zips['zip'] = (ny_zips['zip'].astype(float).astype('Int64').astype(str).str.zfill(5))

train_zip = set(ny_zips.loc[ny_zips['state_id'] == 'NY', 'zip'])
test_zip = set(df['zip_code']) - train_zip
```

```python
train = df[df['zip_code'].isin(train_zip)]
test  = df[df['zip_code'].isin(test_zip)]
```

```python
train.drop(columns=["zip"], inplace=True)
```

```python
test.drop(columns=["zip", "Estimate Married-couple families Mean income (dollars)"], inplace=True)
```
